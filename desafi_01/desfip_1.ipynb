```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
pd.read_csv(url)
```

```python
dados = pd.read_csv(url)
dados.isnull().sum()
```

```python
dados = dados.fillna(0)
dados
dados.isnull().sum()
```

```python
alunos_remover = dados.query('Nome == "Alice" | Nome == "Carlos"').index
dados.drop(alunos_remover, axis=0, inplace=True)
dados
```

```python
selecao = dados['Aprovado'] == True
alunos_aprovados = dados[selecao]
```

```python
alunos_aprovados.to_csv('alunos_aprovados.csv', index=False, sep = ';')
```

```python
pd.read_csv('alunos_aprovados.csv', sep =';')
```

```python
alunos_aprovados = alunos_aprovados.replace(7.0, 8.0)
alunos_aprovados.to_csv('alunos_aprovados.csv', index=False, sep = ';')
pd.read_csv('alunos_aprovados.csv', sep =';')
```

```python
dados = pd.read_csv(url)
```

```python
dados['Pontos_extras'] = dados['Notas'] * 0.4
dados['Pontos_extras'] = dados['Notas'].apply(lambda x: x*0.4)
dados
```

```python
dados['Nota_final'] = dados['Notas'] + dados['Pontos_extras']
dados
```

```python
dados['Aprovado_final'] = dados['Nota_final'].apply(lambda x: 'Aprovado' if x >= 6.0 else 'Reprovado')
dados
```

```python
selecao = (dados['Aprovado'] == False) & (dados['Aprovado_final'] == 'Aprovado')
dados[selecao]
```
