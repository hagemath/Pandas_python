# Tratando e filtrando os dados

## Lidando com dados nulos

```python
df_apartamentos.isnull()
```

```python
df_apartamentos.isnull().sum()
```

```python
df_apartamentos.fillna(0)
```

```python
df = df_apartamentos.fillna(0)
```

```python
df.isnull().sum()
```

## Removendo registros

```python
df.query('Valor == 0 | Condominio == 0')
```

```python
df.query('Valor == 0 | Condominio == 0').index
```

```python
registro_a_remover = df.query('Valor == 0 | Condominio == 0').index
```

```python
df.drop(registro_a_remover, axis=0, inplace = True)
```

```python
df.head(10)
```

```python
df.Tipo.unique()
```

```python
df.drop('Tipo', axis=1, inplace=True)
```

```python
df.head()
```

## Filtros

### **1. Apartamentos que possuem `1 quarto` e `aluguel menor que 1200`**



```python
df['Quartos'] == 1
```

```python
selecao_1 = df['Quartos'] == 1
df[selecao_1]
```

```python
selecao_2 = df['Valor'] < 1200
df[selecao_2]
```

```python
selecao_final = (selecao_1) & (selecao_2)
df[selecao_final]
```

```python
df_1 = df[selecao_final]
```

### **2. `Apartamentos` que possuem pelo menos `2 quartos`, `aluguel menor que 3000` e `area maior que 70`**

```python
ap_quart = df['Quartos'] >= 2
ap_aluguel = df['Valor'] < 3000
ap_area = df['Area'] > 70
```

```python
selecao_final_2 = (ap_quart) & (ap_aluguel) & (ap_area)
df[selecao_final_2]
```

```python
selecao_final_2 = (df['Quartos'] >= 2) & (df['Valor'] < 3000) & (df['Area'] > 70)
df_2 = df[selecao_final_2]
```
