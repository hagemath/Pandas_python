# Análise exploratória de dados

## Qual o valor médio de aluguel por tipo de imóvel?

```python
dados.head(6)
```

```python
dados['Valor'].mean()
```

```python
dados.groupby('Tipo').mean(numeric_only = True) 
```

```python
dados.groupby('Tipo')['Valor'].mean() 
```

```python
dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
mean_preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
mean_preco_tipo.plot(kind='barh', figsize=(6,6), color = 'purple')
```

## Removendo os imóveis comerciais

```python
dados.Tipo.unique()
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala', 
                      'Prédio Inteiro', 'Loja/Salão', 
                      'Galpão/Depósito/Armazém', 
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']
                                            
```

```python
dados.query('@imoveis_comerciais in Tipo')
```

```python
dados.query('@imoveis_comerciais not in Tipo')
```

```python
df_residencias = dados.query('@imoveis_comerciais not in Tipo')
df_residencias.head(10)
```

```python
df_residencias.Tipo.unique()
```

```python
mean_preco_comercial = df_residencias.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
mean_preco_comercial.plot(kind='barh', figsize=(6,6), color='green')
```

## Qual o percentual de cada tipo de imóvel na nossa base de dados?

```python
df_residencias.Tipo.unique()
```

```python
df_residencias.Tipo.value_counts(normalize=True)
```

```python
df_residencias.Tipo.value_counts(normalize=True).to_frame().sort_values('proportion')
```

```python
df_residencias_percentual = df_residencias.Tipo.value_counts(normalize=True).to_frame().sort_values('proportion')
df_residencias_percentual.plot(kind='bar', figsize=(8,6), color='red', xlabel= 'Tipos', ylabel = 'Percentual') 
```

### **Selecionando apenas os imóveis do tipo apartamento**

```python
df_residencias.query('Tipo == "Apartamento"')
```

```python
df_apartamentos = df_residencias.query('Tipo == "Apartamento"')
df_apartamentos.head(10)
```

```python
# 1-) Calcular a média de quartos por apartamento:
df_apartamentos['Quartos'].mean()
```

```python
# 2-) Conferir quantos bairros únicos existem na nossa base de dados
df_completa = pd.read_csv(url, sep=';')
len(df_completa['Bairro'].unique())
```

```python
# 3-)Analisar quais bairros possuem a média de valor de aluguel mais elevadas
df_completa.groupby('Bairro')[['Valor']].mean().sort_values('Valor', ascending=False).head(5)
```

```python
# 4-) Criar um gráfico de barras horizontais que apresente os 5 bairros com as médias de valores de aluguel mais elevadas.
df_bairros_valores = df_completa.groupby('Bairro')[['Valor']].mean().sort_values('Valor', ascending=False).head(5)
df_bairros_valores.plot(kind='barh', figsize=(6,6), color='blue')
```
